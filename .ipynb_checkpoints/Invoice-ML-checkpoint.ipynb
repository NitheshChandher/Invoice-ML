# 1. Text Detection using Google VisionAPI

```python
def extractText(img):
    with io.open(img,'rb') as image_file:
        content= image_file.read()
    image = vision.types.Image(content=content)
    response = client.text_detection(image=image)
    texts = response.text_annotations
    df =pd.DataFrame(columns=['locale','description'])
    for text in texts:
        df= df.append(dict(locale=text.locale,description=text.description), ignore_index= True)
    return df
```

```python
# pdf to image
pages = convert_from_path('invoice.pdf')
for page in pages:
    page.save('invoice.png', 'PNG')
    
# extract text from invoice using VisionAPI
path_to_file = 'invoice.png'
text=pd.DataFrame(columns=['image','text'])
df = extractText(path_to_file)
```
